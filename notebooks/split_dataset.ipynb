```python
import os
label_word = ['a', 'ba', 'dara', 'ei', 'ga', 'ha', 'ka', 'la', 'ma', 'na', 'nga', 'ou', 'pa', 'sa', 'ta', 'wa', 'ya']
split_folder = ['val', 'train']
for split in split_folder:
    for label in label_word:
        path = os.path.join(f"dataset/{split}", label)
        print(path)
        os.mkdir(path)
```

```python
for label in label_word:
    dataset = f"new_dataset/{label}"
    for filename in os.listdir(dataset):
        f = dataset + '/' + filename
        os.replace(f, f"baybayin/{filename}")
```

```python
for split in split_folder:
    dataset = f'dataset_split/{split}/baybayin'
    for filename in os.listdir(dataset):
        f = os.path.join(dataset, filename)
        category = filename.split('_')[0]
        # checking if it is a file
        if os.path.isfile(f):
            os.replace(f, f"dataset/{split}/{category}/{filename}")
            print(f"dataset/{split}/{category}/{filename}")
```
